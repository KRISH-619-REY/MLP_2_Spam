```python
import pandas as pd

df = pd.read_csv("sms+spam+collection/SMSSpamCollection", sep="\t", header=None, names=["label", "message"])
df
```

```python
df.isnull().sum()
```

```python
#Converting all the messages into lower case
df.message=df.message.str.lower()
df.head(10)
```

```python
#Tokenization
import nltk
from nltk.tokenize import word_tokenize
nltk.download("punkt_tab")
```

```python
df["message"]=df["message"].fillna('').apply(word_tokenize)
df.head()
```

```python
df.message = df.message.apply(lambda x : [word for word in x if word.isalpha()])
df.head(10)
```

```python
#Removing all stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
df.message = df.message.apply(lambda x: [word for word in x if not word in stop_words])
df.head(10)
```

```python
#Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
df.message = df.message.apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
df.head(10)
```

```python
#Convert all the token into sentences format
df.message = df.message.apply(lambda x : ''.join(x))
df.head(10)
```

```python
#Apply TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(df["message"]).toarray()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, df["label"], test_size=0.2)
```

```python
#Build MLP model
import tensorflow as tf 
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(xtrain.shape[1],)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
ytrain = le.fit_transform(ytrain)  # "ham"в†’0, "spam"в†’1
ytest = le.transform(ytest)
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(xtrain, ytrain, epochs=20, validation_data=(xtest,ytest), batch_size=128)
```

```python
loss,acc=model.evaluate(xtest,ytest)
```

```python
model.save("spam_model.keras")
```

```python
import tf2onnx
import tensorflow as tf
import onnx

model = tf.keras.models.load_model("spam_model.keras")

@tf.function(input_signature=[tf.TensorSpec(model.inputs[0].shape, model.inputs[0].dtype, name="input")])
def model_fn(x):
    return model(x)

onnx_model, _ = tf2onnx.convert.from_function(
    model_fn,
    input_signature=[tf.TensorSpec(model.inputs[0].shape, model.inputs[0].dtype, name="input")],
    opset=13
)

onnx.save(onnx_model, "spam_model.onnx")
print("Done вњ…")
```
